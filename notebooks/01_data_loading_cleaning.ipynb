```python
data = {
    "IncidentID": range(1, 11),
    "System": [
        "Email", "CRM", "Website", "Database", "Website",
        "Email", "CRM", "Database", "Website", "Email"
    ],
    "Severity": [
        "High", "Medium", "High", "Critical", "Medium",
        "Low", "High", "Critical", "Medium", "Low"
    ],
    "OpenedTime": [
        "2024-01-01 08:15", "2024-01-02 09:30", "2024-01-03 11:00",
        "2024-01-04 14:45", "2024-01-05 10:20", "2024-01-06 16:00",
        "2024-01-07 13:10", "2024-01-08 18:30", "2024-01-09 09:50",
        "2024-01-10 07:40"
    ],
    "ResolvedTime": [
        "2024-01-01 10:45", "2024-01-02 12:00", "2024-01-03 15:30",
        "2024-01-04 20:15", "2024-01-05 12:00", "2024-01-06 17:10",
        "2024-01-07 16:40", "2024-01-08 23:30", "2024-01-09 12:30",
        "2024-01-10 09:00"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df["OpenedTime"] = pd.to_datetime(df["OpenedTime"])
df["ResolvedTime"] = pd.to_datetime(df["ResolvedTime"])

df.dtypes
```

```python
df["ResolutionMinutes"] = (
    df["ResolvedTime"] - df["OpenedTime"]
).dt.total_seconds() / 60

df[["IncidentID", "System", "Severity", "ResolutionMinutes"]]
```
